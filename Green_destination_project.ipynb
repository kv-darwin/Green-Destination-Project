```python
plp_lft = data['Attrition'].sum()
```

```python
plp_lft
```

```python
tot_plp = len(data)
```

```python
tot_plp
```

```python
plp_rate = (plp_lft/tot_plp)*100
```

```python
plp_rate
```

```python
age_corr = data['Age'].corr(data['Attrition'])
age_corr
```

```python
year_corr = data['YearsAtCompany'].corr(data['Attrition'])
year_corr
```

```python
income_corr = data['DailyRate'].corr(data['Attrition'])
income_corr
```

```python
print("Attrition Rate:{:.2f}%".format(plp_rate))
```
